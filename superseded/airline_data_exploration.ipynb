### State Level and Airline Level Calculations

 * As the dataset is very large it makes sense to take a subset of the data to work with.
 * Split for a single state (largest - CA & TX have 600k flights per year = around 12% of dataset each)
 * Split for a single state (medium size @ the 50,000 flights in 2015 level)
 * Split by Airline

```python
state_totals = df.groupby(['ORIGIN_STATE']).ORIGIN_STATE.agg('count')
state_totals.head()
```

```python
state_totals
```

```python
airline_totals = df.groupby(['AIRLINE']).AIRLINE.agg('count')
airline_totals.head()
```
